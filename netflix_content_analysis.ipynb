## Data Cleaning

The dataset contains some missing values.  
Before visualization, the data needs basic preprocessing and formatting.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')

df['year_added'] = df['date_added'].dt.year

df.head()
```

## Movies vs TV Shows

```python
import matplotlib.pyplot as plt

type_counts = df['type'].value_counts()

plt.figure(figsize=(6,4))
type_counts.plot(kind='bar')

plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')

plt.xticks(rotation=0)

plt.show()
```

## Top 10 Countries Producing Netflix Content

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind='bar')

plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')

plt.xticks(rotation=45)

plt.show()
```

## Content Release Trends

```python
release_years = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,5))
release_years.plot()

plt.title('Netflix Content Release Trend')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.show()
```
